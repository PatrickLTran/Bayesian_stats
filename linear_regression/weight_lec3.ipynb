# Estimate weight given height of patient
Assume normal distribution where H_bar is avg height

alpha becomes the average weight when H_i = H_bar 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#y = a + bx

n_samples = 10
alpha = np.random.normal(loc=0,scale=1.0, size=n_samples)
beta = np.random.normal(loc=0,scale=1.0, size=n_samples)

x = range(0, 10)
plt.scatter(alpha, beta)
for beta, alpha in zip(beta, alpha):
    y = [beta*i + alpha for i in x]  
    plt.plot(x, y, label=f'y = {beta}x + {alpha}')  

plt.legend() 
plt.show() 
```
